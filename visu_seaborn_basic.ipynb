```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
fmri=sns.load_dataset('fmri') # buit in data set in seaborn
```

```python
fmri.head()
```

# line plot

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
x=fmri.loc[:,('timepoint')]

y=fmri.loc[:,('signal')]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

# bar plot

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon)
plt.show()
```

# scatter plot

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.columns
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species',style='Species',marker=True)
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='SepalWidthCm',style='Species',marker=True)
plt.show()
```

# histogram or distribution plot

```python
diamonds=pd.read_csv('diamonds.csv')
```

```python
diamonds.columns
```

```python
diamonds.dtypes
```

```python
diamonds.head()
```

```python
sns.distplot(diamonds['price']) # histogram + frequency polygon
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False,color='g') # no histogram only frequency curve
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red')
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red',kde=False,bins=500)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red',vertical=True)
plt.show()
```

# joint plot (scatter plot + histogram)

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species')
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,kind='reg')
plt.show()
```

# Box plot

```python
churn=pd.read_csv('churn.csv')
```

```python
churn.columns
```

```python
churn.dtypes
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,color='g')
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,linewidth=5)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,linewidth=5,order=['no','DSL','Fiber optics InternetService'])
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,linewidth=2,hue='PaymentMethod')
plt.show()
```
