# line plot

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,11)
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x
```

```python
y
```

```python
z=np.random.randint(1,20,10)
z
```

```python
plt.plot(x,y,color='g',linewidth=8)
plt.plot(x,z)
plt.title('line plot')
plt.xlabel('gujhuiu')
plt.ylabel('fynm')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='g',linestyle=':',linewidth=2)

plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,z,color='r',linestyle=':',linewidth=2)

plt.grid(True)

plt.show()
```

# bar plot (cat vs num)

```python
student={'a':10,'b':25,'c':20,'d':35}
```

```python
names=list(student.keys())
names
```

```python
values=list(student.values())
values
```

```python
plt.bar(names,values,color='y')
plt.title('bar plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.scatter(x,z,c='y',marker='*',s=100)
plt.scatter(x,y,c='r',s=100)
plt.title(' plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
plt.hist(z)
plt.show()
```
